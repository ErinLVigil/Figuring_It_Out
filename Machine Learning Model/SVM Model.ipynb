```python
# Create the OneHotEncoder instance
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
df3 = pd.DataFrame(enc.fit_transform(df.Type.values.reshape(-1,1)))

# Rename encoded columns
df3.columns = enc.get_feature_names(['Type'])
df3.head()
```

```python
merged_df = pd.DataFrame(df.merge(df3,left_index=True,right_index=True).drop("Type",1))

```

```python
merged_df
```

```python
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
emp_df = pd.DataFrame(enc.fit_transform(merged_df.Employee_Count.values.reshape(-1,1)))

# Rename encoded columns
emp_df.columns = enc.get_feature_names(['Employee_Count'])
emp_df.head()
```

```python
merged2_df = pd.DataFrame(merged_df.merge(emp_df,left_index=True,right_index=True).drop("Employee_Count",1))
```

```python
merged2_df
```

```python
enc = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
sales_df = pd.DataFrame(enc.fit_transform(merged2_df.Annual_Sales.values.reshape(-1,1)))

# Rename encoded columns
sales_df.columns = enc.get_feature_names(['Annual_Sales'])
sales_df.head()
```

```python
merged3_df = pd.DataFrame(merged2_df.merge(sales_df,left_index=True,right_index=True).drop("Annual_Sales",1))
```

```python
merged3_df
```

```python
# Remove loan status target from features data
y = merged3_df.successful.values
X = merged3_df.drop(columns=['successful']).values

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create the SVM model
svm = SVC(kernel='linear')
```

```python
# Train the model
svm.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = svm.predict(X_test_scaled)
print(f" SVM model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```
