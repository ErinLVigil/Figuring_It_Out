```python
import pandas as pd

census_df = pd.read_csv('population_by_zip_2010.csv')

census_df
```

```python
census_new = census_df.sort_values('population', ascending=False).drop_duplicates('zipcode').sort_index()
```

```python
census_new
```

```python
census_csv = census_new.drop(columns=['minimum_age', 'maximum_age', 'gender', 'geo_id'])
```

```python
census_csv
```

```python
census_csv.to_csv('census_2010.csv', index=False)
```
